# この本で学ぶこと
* 統計モデリング（モデルを作って観測データに当てはめて現象を理解する方法）について理解する
* 統計モデルのひとつである`一般化線形モデル`とその発展について学ぶ

# 1. データを理解するために統計モデルを作る

# 1.1 統計モデルとは
統計モデルとは、以下の特徴をもつ数理モデルのことです。
* 観測によってデータ化された現象を説明するために作られる
* 確率分布を基本的な部品とする
* データとモデルを対応づける手続きが用意されている

## 統計モデルによって、現象を理解する
自然の仕組みを理解しようとする時、人間が解釈を行うまでに2段階の情報消失が発生すると考えられます。  
第一段階では、観測や実験といった手段で対象から記号の集まりである「観測データ」に変化します。  
第二段階では、観測データを解析（統計モデルを用いた要約や整理）をすることによって何らかの解釈を行います。  
  
上記の2段階に置いて、情報の消失が起こります。  
こうした情報の消失には「ばらつき」、「欠測」、「誤差」などがありますが、統計モデルの部品である確率分布によってこれらをうまく扱うことができます。

# 1.2 「ブラックボックス統計学」
中身を理解せずに統計手法を用いるのは**疑似科学**の作法なのでやめましょう。  
例えば、以下のような状況がそうです。
* **ゆうい差**がでるまで検定手法をひたすら取り替える
* R^2値は「説明力」なので、ひたすら1に近ければよい

このような疑似科学の手法を本の著者は「ブラックボックス統計学」と呼び、避けるべきだと説いています。  
そのために必要なこととして、解析者がデータを見て目的に添いつつ、そのデータ構造に合致した統計モデル構築できるようになること（この本の目的）が重要だということです。
  
（ちなみにノートブックの作者はこのようなブラックボックス統計学のことを、統計ソフト使って中身を理解せずにぽちぽちやる様子から**ワンクリック学派**と呼んでいます。ちなみに出展は「読む統計学 使う統計学（著：広田すみれ）」です。）

# 1.3 本の内容（GLMの位置付け）
回帰や分散分析等の手法と**一般化線形モデル（generalized linear model, GLM）**の関係を整理します。  
ほとんどの統計ソフトウェアに付いている分析手法は、データのばらつきについて、等分散正規分布が仮定されています。  
そうした正規分布を前提とした統計モデルは**線形モデル**と呼ばれ、より一般的な確率分布へ拡張した統計モデルが**一般化線形モデル（GLM）**となります。
